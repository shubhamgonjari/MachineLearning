# Salary_data Simple Linear Regression

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv(r'C:\Users\Administrator\Desktop\ml_prac\Salary_Data.csv')
data.head()
```

```python
sns.set()
sns.scatterplot(x='YearsExperience',y='Salary',data=data)
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(data.YearsExperience,data.Salary)
```

```python
r_value ** 2
```

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitLine = predict(data.YearsExperience)

plt.scatter(data.YearsExperience, data.Salary)
plt.plot(data.YearsExperience, fitLine, c='r')
plt.show()
```

## Sklearn Approch

```python
from sklearn.model_selection import train_test_split
X=data['YearsExperience']
y=data['Salary']
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.30,random_state=2018)

import numpy as np
X_train=X_train[:,np.newaxis]
X_test=X_test[:,np.newaxis]
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print(mean_squared_error(y_test,y_pred))
```

```python
print(r2_score(y_test,y_pred))
```

```python
print(lr.intercept_)
```

```python
print(lr.coef_)
```

## Fitting line to Training set

```python
plt.scatter(X_train,y_train,color='r')
plt.plot(X_train,lr.predict(X_train),color='green')
plt.title('Salary VS Years of Experience(Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Predicting on Test Set

```python
plt.scatter(X_test,y_test,color='r')
plt.plot(X_train,lr.predict(X_train),color='green')
plt.title('Salary VS Years of Experience(Testing set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
