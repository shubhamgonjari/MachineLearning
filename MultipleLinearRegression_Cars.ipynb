```python
sig_col = ['price','enginetype','fueltype', 'aspiration','carbody','cylindernumber', 'drivewheel',
            'wheelbase','curbweight', 'enginesize', 'boreratio','horsepower', 
                    'citympg','highwaympg', 'carlength','carwidth']
```

```python
data=data[sig_col]
```

```python
sig_cat_col=['fueltype','aspiration','carbody','drivewheel','enginetype','cylindernumber']
```

```python
dummy=pd.get_dummies(data[sig_cat_col],drop_first=True)
dummy.shape
```

```python
data = pd.concat([data, dummy], axis = 1)
```

```python
data.drop(sig_cat_col,inplace=True,axis=1)
```

```python
data.shape
```

```python
X=data.iloc[:,1:]
y=data.iloc[:,0]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=2018)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
print(r2_score(y_test,y_pred))
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```
